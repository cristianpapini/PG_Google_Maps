```python
import polars as pl
import pandas as pd

import numpy as np
import panel as pn
pn.extension('tabulator')
from panel.interact import interact

import hvplot.pandas

import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
```

## Abro fichero

```python
pl_lugares = (
    pl.read_parquet(r"lugares.parquet")
    .select(pl.exclude(['closed']))
)
pl_lugares.head()
```

## Creo un pandas dataframe

```python
pd_lugares = pl_lugares.to_pandas()
```

## Creo el interactive

```python
i_lugares = pd_lugares.interactive()
```

## Creo el widget para editar la categoria

```python
edit_lugares = pn.widgets.TextInput(name='Category (en Mayúsculas)',value='')
edit_lugares
```

```python
edit_ciudad = pn.widgets.TextInput(name='Ciudad (en Mayúsculas)',value='')
edit_ciudad
```

## Creo el pipeline pipe_lugares a partir del interactive i_lugares
### En la declaracion estoy involucrando al widget edit_lugares

```python
pipe_lugares = (
    i_lugares[
    i_lugares['City'].apply(lambda x: x.upper()).str.contains(edit_ciudad) &
    i_lugares['category'].apply(lambda x: x.upper()).str.contains(edit_lugares)    
    ][['City','name','address','category']]    
    .reset_index(drop=True)
)
#pipe_lugares.head()
```

## Conecto el pipeline pipe_lugares con un widget
### Y voy a tener en la visualizacion al widget edit_lugares por estar involucrado en la
### definicion del pipeline
## El widget creado se llama vista_lugares

```python
vista_lugares = pipe_lugares.pipe(pn.widgets.Tabulator, pagination = 'remote', page_size = 10, sizing_mode='stretch_width')
vista_lugares
```
